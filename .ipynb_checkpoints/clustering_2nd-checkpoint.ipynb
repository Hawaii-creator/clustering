```python
"""
clustering_1stで作成したkmeansモデルを使って、他のwineデータもクラスタリングする
"""
```

```python
# 必要なライブラリーのインストール
import numpy as np
import pandas as pd
import sys
import pickle
from pickle import load
# 可視化
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
%matplotlib inline
sns.set_style('whitegrid')

# 正規化のためのクラス
from sklearn.preprocessing import StandardScaler

# k-means法に必要なものをインポート
from sklearn.cluster import KMeans
from sklearn import datasets, preprocessing
```

```python
data = datasets.load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
df =df.loc[:,["alcohol","color_intensity", "proline"]]
```

```python
#データを70件、重複を許可して抽出
df_B = df.sample(n=70, replace=True)
```

```python
df_B.shape
```

```python
#clustering_1stで保存したpickleからロードして標準化
for i in df_B.columns:
    scaler = load(open(r"pickle/pickle_A_{0}.pkl".format(i), "rb"))
    df_B[i] = scaler.transform(df_B[[i]].values)
df_B
```

```python
#clustering_1stで保存したpickleからロードしてkmeans
kmeans = load(open(r"pickle/cls_A.pkl", "rb"))
```

```python
kmeans_record = kmeans.predict(df_B)
```

```python
len(kmeans_record)
```

```python
df_B["cluster"] = kmeans_record
df_B.head()
```

```python
#クラスタごとの数をカウント
df_B_groupby = df_B.groupby("cluster").count()
df_B_groupby
```
