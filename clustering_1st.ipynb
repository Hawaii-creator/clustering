```python
# 必要なライブラリーのインストール
import numpy as np
import pandas as pd
import sys
import pickle

# 可視化
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
%matplotlib inline
sns.set_style('whitegrid')

# 正規化のためのクラス
from sklearn.preprocessing import StandardScaler

# k-means法に必要なものをインポート
from sklearn.cluster import KMeans
from sklearn import datasets, preprocessing
```

```python
data = datasets.load_wine()
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df =df.loc[:,["alcohol","color_intensity", "proline"]]
```

```python
df.shape
```

```python
#データを100件、重複を許可して抽出
df_A = df.sample(n=100, replace=True)
```

```python
df_A.shape
```

```python
for i in df_A.columns:
    print(i)
```

```python
#標準化
sc = StandardScaler()
for i in df_A.columns:
    df_A[i] = sc.fit_transform(df_A[[i]].values)
    #scをpickleとして保存
    pickle.dump(sc, open(r"pickle/pickle_A_{0}.pkl".format(i), "wb"))
```

```python
cls = KMeans(n_clusters=3, random_state=42)
#クラスタリング実行
kmeans_model = cls.fit(df_A)
```

```python
#クラスタを付与
df_A["cluster"] = kmeans_model.labels_
df_A.head()
```

```python
#kmeansをpickleとして保存
pickle.dump(cls, open(r"pickle/cls_A.pkl", "wb"))
```

```python
#クラスタごとの数をカウント
df_A_groupby = df_A.groupby("cluster").count()
df_A_groupby
```
